```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

base_url = 'http://house.jumia.cm/apartment/rent/?page='

urls = []
for i in range(1, 186):
    url = base_url + str(i) + '&size='
    urls.append(url)

my_houses = []
my_adds = []
my_prices = []
my_rooms = []
my_bathrooms = []

for link in urls:
    code = requests.get(link)
    soup = BeautifulSoup(code.content)
    print 'Retrieving ' + link
    print '.............................................'
    
    houses = [item.get_text() for item in soup.findAll('h3')]
    for h in houses:
        my_houses.append(h)
    
    addresses= [item.get_text() for item in soup.findAll('p', {"class": "listing-address icon-location"})]
    for ad in addresses:
        my_adds.append(ad)
   
    prices = [item.get_text() for item in soup.findAll('span', {"class": "listing-price"})]
    for p in prices:
        my_prices.append(p)
        
                
    for at in soup.find_all('ul', {"class": "listing-attributes"}):
        
        try:
            all_rooms = at.text.strip()
            rooms1 = all_rooms.split('\n')
            rooms = rooms1[0]
            my_rooms.append(rooms)
            baths = rooms1[4]
            my_bathrooms.append(baths)
            
        except:
            my_bathrooms.append("NA")
            continue



print 'DONE!'

    
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

base_url = 'http://www.koutchoumi.com/en/main/showResults?page='

urls = []

for i in range(1, 260):
    url = base_url + str(i) + '&k[typeBien]=1&k[typeTransaction]=0&k[villeId]=-1&k[prixMin]=&k[prixMax]=&k[surfaceMin]=&k[surfaceMax]=&k[nombreChambres]=-1&k[quartierId]=&k[minSDB]=&k[maxSDB]=&k[avecParking]=&k[avecBarriere]=&k[typeMaisonId]=&k[situationSalleDeBains]=&k[situationToilettes]=&k[titre]=&k[distanceRouteMin]=&k[distanceRouteMax]=&k[avecDouche]=&k[nombrePiecesMin]=&k[nombrePiecesMax]='
    
    urls.append(url)

count = 0

types = []
bedrooms = []
areas = []
bathrooms = []
all_prices = []
    
for link in urls:
    code = requests.get(link)
    soup = BeautifulSoup(code.content)
    
    count += 1
    print 'Retrieving page ' + str(count)
#    house_data = soup.find_all('div', {"class": "content_result"})
    print '.............................................'
#    print house_data[0]

    house =[item.get_text() for item in soup.findAll("h1", {"class": "title_blue"})]
    for el in house:
        ho = el.split('-')    
    
        house_type = ho[0].strip()
        types.append(house_type)
    
        area = ho[1].strip()
        areas.append(area)
    
    for i in soup.find_all('div', {"class": "content_result"}):
        rooms = i.find_all('li')[2].get_text()
#        for r in rooms:
        bedrooms.append(rooms)
    
        baths = i.find_all('li')[3].get_text()
#    for b in baths:
        bathrooms.append(baths)
    
        price = i.find_all('li')[0].get_text()
#    for p in price:
        all_prices.append(price)
        
#print types  
#print areas
#print bedrooms
```
