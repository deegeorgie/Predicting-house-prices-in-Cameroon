```python
import requests
from bs4 import BeautifulSoup

base_URL = 'http://house.jumia.cm/house/rent/?page='

urls = []

for page in range(1, 18):
    url = base_URL + str(page) + '&size='
    urls.append(url)
    
types = []
areas = []
prices = []
bedrooms = []
bathrooms = []
for link in urls:
    code = requests.get(link)
    soup = BeautifulSoup(code.content)
#    print 'Retrieving ' + link
#    print '.............................................'

    house_type = [item.get_text() for item in soup.find_all('h3')]
    for t in house_type:
        h = t.strip()
        types.append(h)
    
    address = [item.get_text() for item in soup.find_all('p', {'class': 'listing-address icon-location'})]
    for ad in address:
        area = ad.strip()
        areas.append(area)
    
    price = [item.get_text() for item in soup.find_all('span', {'class': 'listing-price'})]
    for p in price:
        lp = p.replace(',', ' ')
        prices.append(lp)
        
    for at in soup.find_all('ul', {"class": "listing-attributes"}):
        
        try:
            all_rooms = at.text.strip()
            rooms1 = all_rooms.split('\n')
            rooms = rooms1[0]
            bedrooms.append(rooms)
            baths = rooms1[4]
            bathrooms.append(baths)
            
        except:
            bathrooms.append("NA")
            continue
print bathrooms

        
  
```

```python
import pandas as pd

other_houses = pd.DataFrame({
        'Area': areas,
        'Description': types,
        'Bedrooms': bedrooms,
        'Bathrooms': bathrooms,
        'Price': prices
    })
```
